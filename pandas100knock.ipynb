```python
#90 df_copyのsexとembarked列をラベルエンコーディング
df = pd.read_csv('https://raw.githubusercontent.com/jbryer/CompStats/master/Data/titanic3.csv')
df_copy = df.copy()

le = LabelEncoder()
df_copy['embarked'] = le.fit_transform(df_copy['embarked']) 
df_copy['sex'] = le.fit_transform(df_copy['sex']) 
df_copy.head()
```

```python
#91 df_copyの欠損値を確認
df_copy.isnull()
df_copy.isnull().sum()
```

```python
#92 df_copyのage、fare列の欠損値を各列の平均値で補完
df_copy['age'] = df_copy['age'].fillna(df_copy['age'].mean())
df_copy['fare'] = df_copy['fare'].fillna(df_copy['fare'].mean())
df_copy.isnull().sum()
```

```python
#93 df_copyの中で機械学習で使用しない不要な行を削除
df_copy = df_copy.drop(['name', 'ticket', 'cabin', 'boat', 'body', 'home.dest'],axis=1)
df_copy
```

```python
#94 ①df_copyのpclass、age、sex、fare、embarkedの列を抽出し、ndarray形式に変換②df_copyのsurvivedの列を抽出し、ndarray形式に変換
features = df_copy[['pclass','age','sex','fare','embarked']].values
target = df_copy['survived'].values
```

```python
#95 【94】で作成したfeatrues、targetを学習データとテストデータに分割
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.33, random_state=42)
```

```python
#96 学習データ(features、target)を用いランダムフォレストにて学習を実行
from sklearn.ensemble import RandomForestClassifier

model =  RandomForestClassifier(max_depth=30, n_estimators=30, random_state=42)

model.fit(X_train,y_train)
```

```python
#97 test_Xデータの乗客の生存を予測
y_pred=model.predict(X_test)
```

```python
#98 予測結果がtest_y(生存有無の答え)とどれぐらい整合していたかを確認(評価指標はaccuracy)
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
#99 学習における各列(特徴量)の重要度を表示
importace = model.feature_importances_ 

print('Feature Importances:')
for i, feat in enumerate(['pclass','age','sex','fare','embarked']):
    print('\t{0:20s} : {1:>.5f}'.format(feat, importace[i]))
```

```python
#100 test_Xの予測結果をcsvでoutputフォルダに出力（ファイル名は「submission.csv」）
df_pred = pd.DataFrame(y_pred)
df_pred.to_csv('output/submission.csv',header=None)
```
